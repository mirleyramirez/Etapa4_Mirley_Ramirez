```python
import pandas as pd
```

```python
data = pd.read_csv('Cleaned-Data.csv',
engine='python', index_col=0)
```

```python
data.head()
```

```python
data.info()
```

```python
X=data.iloc[:,1:24]
Y=data.iloc[:,0]
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.75, random_state=0)
```

```python
X_train.info()
```

```python
from sklearn.tree import DecisionTreeClassifier
arbol=DecisionTreeClassifier(max_depth=4)
arbol_covid19 = arbol.fit(X_train, Y_train)
```

```python
from matplotlib import pyplot as plt
from sklearn import tree

fig = plt.figure(figsize=(25, 20))
class_names = list(map(str, Y.values))
tree.plot_tree(arbol_covid19, feature_names=list(X.columns.values),
               class_names=class_names, filled=True)
plt.show()
```

```python
fig.savefig("Arbol de desicion.png")
```

```python
Y_predic=arbol_covid19.predict(X_test)
Y_predic
```

```python
from sklearn.metrics import confusion_matrix
Matriz_de_confusion=confusion_matrix(Y_test, Y_predic)
Matriz_de_confusion
```

```python
import numpy as np
precision_Global=np.sum(Matriz_de_confusion.diagonal())/np.sum(Matriz_de_confusion)
precision_Global
```
